```
# 만료연도가 25인 것만 추출하기
ecom[ecom['CC Exp Date'].apply(lambda x: x[3:]) == '25'].head(5)
```

```
# 끝자리 25인지 확인 
sum(ecom['CC Exp Date'].apply(lambda x: x[3:] == '25'))
```
